```python
import sqlite3
import matplotlib.pyplot as plt
import pandas as pd

# Connect to the database
conn = sqlite3.connect('BenchmarkSubsetIterationsCDF.db')

# Specify the algorithm name to filter by

modes = ["walksat", "coloringA1_heuristic0", "coloringA1_heuristic1",
             "coloringA1_heuristic2", "coloringA1_heuristic3"]

algorithm_name = "coloringA1_heuristic0"

inside = """
            filename TEXT,
            algorithmname TEXT,
            avg_flips REAL,
            prob_s REAL,
            std_flips REAL,
            tts_99 REAL,
            avg_loops REAL,
            std_loops REAL,
            avg_tries REAL,
            std_tries REAL

"""

query = """
SELECT avg_loops, prob_s 
FROM BenchmarkSubsetIterationsCDF
WHERE algorithmname = ?
ORDER BY avg_loops;
"""

# Load the data into a DataFrame
df = pd.read_sql_query(query, conn, params=(algorithm_name,))

# Close the database connection
conn.close()

# Check if data is loaded and then plot
if df.empty:
    print("No data found for this query.")
else:
    # Plot prob_s vs avg_flips
    plt.scatter(df['avg_loops'], df['prob_s'], marker='o')
    plt.xlabel('Average Loops (avg_loops)')
    plt.ylabel('Probability of Success (prob_s)')
    plt.title(f'Probability of Success vs Average Tries for {algorithm_name}')
    plt.grid(True)
    plt.show()
```

```python
import sqlite3
import matplotlib.pyplot as plt
import pandas as pd

# Connect to the database
conn = sqlite3.connect('BenchmarkSubsetIterationsCDF.db')

# Query the data for a specific algorithm, e.g., "Algorithm1"
algorithm_name = "Algorithm1"  # Replace with the desired algorithm name
query = f"""
SELECT avg_loops, prob_s 
FROM BenchmarkSubsetIterationsCDF 
WHERE algorithmname = ?
ORDER BY avg_loops;
"""

# Load the data into a DataFrame
df = pd.read_sql_query(query, conn, params=(algorithm_name,))

# Close the database connection
conn.close()

# Plot prob_s vs avg_loops
plt.plot(df['avg_loops'], df['prob_s'], marker='o')
plt.xlabel('Average Loops (avg_loops)')
plt.ylabel('Probability Success (prob_s)')
plt.title(f'Probability Success vs Average Loops for {algorithm_name}')
plt.grid(True)
plt.show()
```
